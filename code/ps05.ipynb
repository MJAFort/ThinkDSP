```python
response = thinkdsp.read_wave('180960__kleeb__gunshot.wav')
response = response.segment(start=0.26, duration=5.0)
response.normalize()
response.plot()
thinkplot.config(ylim=[-1.05, 1.05],label = False)
```

```python
response.make_audio()
```

```python
transfer = response.make_spectrum()
transfer.plot()
```

```python
transfer.plot()
thinkplot.config(xscale='log', yscale='log')
```

```python
saxophone = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav')
saxophone.ys = saxophone.ys[:len(response)]
saxophone.normalize()
saxophone.plot()
thinkplot.config(ylim=[-1.05, 1.05])
```

```python
saxophone.make_audio()
```

```python
spectrum = saxophone.make_spectrum()
len(spectrum.hs), len(transfer.hs)
output = (spectrum * transfer).make_wave()
output.normalize()
```

```python
saxophone.plot()
output.plot()
```
