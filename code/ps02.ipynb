# 1.

```python
signal = thinkdsp.SinSignal(freq=440)
```

```python
duration = signal.period * 30.5
wave = signal.make_wave(duration)
```

```python
spectrum = wave.make_spectrum()
```

```python
spectrum.plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

Here is the plot of the sine signal without any windowing.

```python
wave.hamming()
spectrum = wave.make_spectrum()
spectrum.plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

Here we have the spectrum after the hamming window was applied.

```python
signal = thinkdspHanning.SinSignal(freq=440)
```

```python
duration = signal.period * 30.5
wave = signal.make_wave(duration)
```

```python
wave.hanning()
spectrum = wave.make_spectrum()
spectrum.plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

A hanning window applied to the same signal.

# 4.

The following recording is from freesound.org and is a recording of someone saying the phonetic vowel sounds.  Ipython notebook thought it would be hilarious to play everything at half speed.

```python
vowels = thinkdsp.read_wave('vowels.wav')
vowels.make_audio()
```

```python
spectrogram = vowels.make_spectrogram(seg_length=1024)
spectrogram.plot(high=64)
thinkplot.config(xlabel='time(s)', ylabel='frequency (Hz)', legend=False)
```

This is the spectrogram of the entire wave.  The recording goes: aeiou.

```python
start = 1.5
duration = 1.25
segmenta = vowels.segment(start,duration)
spectrogram = segmenta.make_spectrogram(seg_length=1024)
spectrogram.plot(high=64)
thinkplot.config(xlabel='time(s)', ylabel='frequency (Hz)', legend=False)
```

Above is the spectrogram of the vowel a.

```python
start = 6.8
duration = 1
segmenti = vowels.segment(start,duration)
spectrogram = segmenti.make_spectrogram(seg_length=1024)
spectrogram.plot(high=64)
thinkplot.config(xlabel='time(s)', ylabel='frequency (Hz)', legend=False)
```

We contrast the spectrogram of the vowel a with the spectrogram for the vowel i and see that i has considerably lower frequency over time.

# 5.

The first wave is a struck tibetan singing bowl.

```python
start = 0
duration = 5
chime = thinkdsp.read_wave('chime.wav')
chimeSegment = chime.segment(start, duration)
chimeSegment.make_audio()
```

```python
chimeSpectrum = chimeSegment.make_spectrum()
chimeSpectrum.plot()
```

```python
chimeSpectrum.plot_power(low=1, linewidth=1, alpha=0.5)
thinkplot.config(xlabel='frequency (Hz)',
                 ylabel='power density',
                 xscale='log',
                 yscale='log',
                 legend=False)
```

```python
chimeSpectrum.estimate_slope()
```

The power spectrum reveals that the slope is approximately -1.  This indicates that the waveform is more closely related to pink noise than to white or brown.

Our second wave is a tibetan singing bowl that is played by scraping across the top of the bowl.
