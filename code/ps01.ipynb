```python
start = 5.6
duration = .01
segment = wave.segment(start, duration)
```

```python
spectrum.low_pass(1000)
```

```python
cos_sig = thinkdsp.CosSignal(freq=440, amp=1.0, offset=0)
```

```python
wave1 = cos_sig.make_wave(duration=0.5, start=0, framerate=11025)
```

```python
wave1.make_audio()
```

I began by synthesizing a cosine wave with 440Hz and listening.

```python
cos_sig1 = thinkdsp.CosSignal(freq=880, amp=1.0, offset=0)
cos_sig2 = thinkdsp.CosSignal(freq=1760, amp=1.0, offset=0)
```

```python
mix=cos_sig+cos_sig1+cos_sig2
```

```python
mixWave1 = mix.make_wave(duration=0.5, start=0, framerate=11025)
```

```python
mixWave1.make_audio()
```

I added in two more harmonics just to be sure of what harmonics sounded like.

```python
cos_nonHarm = thinkdsp.CosSignal(freq=1000, amp=1.0, offset=0)
```

```python
new_mix = mix+cos_nonHarm
```

```python
new_mix_wave = new_mix.make_wave(duration=.5, start=0, framerate=11025)
```

```python
new_mix_wave.make_audio()
```

```python
cos_nonHarm1 = thinkdsp.CosSignal(freq=14100, amp=1.0, offset=0)
cos_nonHarm2 = thinkdsp.CosSignal(freq=700, amp=1.0, offset=0)
new_cos=cos_nonHarm1+cos_nonHarm2
```

```python
new_mix1=new_mix+new_cos
```

```python
new_mix_wave1=new_mix1.make_wave(duration=.5, start=0, framerate=11025)
```

```python
new_mix_wave1.make_audio()
```

As I add in waves that are not harmonics of the beginning frequency, instead of changing the pitch of the tone, I begin to hear an overlaying of different tones.  This secondary tone indicates a change of timbre, and sounds distinctly less clear.
