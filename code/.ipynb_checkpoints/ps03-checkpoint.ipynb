```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import thinkstats2
import math
import numpy
import pandas

%precision 3
%matplotlib inline

PI2 = 2 * math.pi
```

# 1.

```python
nrows = 1625
df = pandas.read_csv('coindesk-bpi-USD-close.csv', nrows=nrows, parse_dates=[0])
ys = df.Close.values
bitwave = thinkdsp.Wave(ys, framerate=1)
bitwave.plot()
```

```python
bitSpectrum = bitwave.make_spectrum()
bitSpectrum.plot_power(low=1, linewidth=1, alpha=0.5)
thinkplot.config(xlabel='frequency (Hz)',
                 ylabel='power density',
                 xscale='log',
                 yscale='log',
                 legend=False)
```

```python
bitSpectrum.estimate_slope()
```

The power spectrum shows that the bitCoin noise most closely resembles red/brownian noise, but is a little pinker than that as the slop is -1.8 rather than -2 like brownian noise.

# 2.

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr
```

```python
serial_corr(bitwave)
```

The serial correlation is about 1 which is expected of red noise.  The autocorrelation is high enough to more closely resemble red noise than the power spectrum would have suggested.

```python
def autocorr(wave):
    lags=range(len(wave.ys)//2)
    corrs=[serial_corr(wave, lag) for lag in lags]
    return lags, corrs

lags, corrs = autocorr(bitwave)
thinkplot.plot(lags,corrs)


```

```python
lags, corrs = autocorr(bitwave)
thinkplot.plot(lags,corrs)
thinkplot.config(xlabel='lag',
                 ylabel='correlation',
                 xscale='log',
                 yscale='log',
                 legend=False)
```
