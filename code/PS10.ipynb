### System characterization

The system is described by the transfer function of series RC or series RLC circuit: $$H(s)=\frac{1/RC}{s+1/RC} \qquad \text{or} \qquad H(s)=\frac{1/LC}{s^2+(R/L)s+1/LC}$$

```python
C=1e-7
L=1e-2
R=50#400 or 50 ohm
```

# Question 3

## Characterization (A)

```python
sys = signal.lti([1,0], [1,1])
sys_string='A'#identifier for graphs
```

```python
time_scale=5*L/R#how long graphs should go for; scales inputs accordingly
t=np.linspace(0,10*L/R,500)
```

```python
_t,s=signal.step(sys,T=t)
timeplot(t,s,label=sys_string+' step response (V)')
```

The step response of the system shows that the system impulse response is flat and therefore is a highly stable system. There is no oscillation which indicates that the pole-zero map has no imaginary components.

```python
import decimal
decimal.getcontext().prec=3
res_freq=decimal.Decimal(1/np.sqrt(L*C)*1/(2*pi))


freqs=np.logspace(3,5,500)
_w, mag, phase = signal.bode(sys,w=2*pi*freqs)#Bode takes frequencies in rad/s and returns magnitudes in dB
logplot(freqs,10**(mag/20),phase,title=sys_string+' Bode plot',margins=0.4)
```

```python
pzmap(sys)
```

From the pole zero plot we can see that there is no imaginary part of the system.  Therefore, there is no oscillation and the system forms an exponential. The overall system is a first order system with one pole and one zero. There is no imaginary part so there are no oscillations. The pole and zero are both real.
